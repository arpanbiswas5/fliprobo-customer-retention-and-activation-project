```python
a = df['From the following, tick any (or all) of the online retailers you have shopped from;                                                                           '].str.get_dummies(',')
sn.barplot(x=a.sum(),y=a.sum(),hue=a.columns,dodge=False)
plt.xlabel("From the following, tick any (or all) of the online retailers you have shopped from")
plt.title("Respondents shopping website list", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sn.countplot(y=df['Easy to use website or application'],orient='h')
plt.show()

df['Easy to use website or application'].value_counts()
```

```python

a = df['Complete, relevant description information of products'].str.get_dummies(',')
a["flipkart.com"] = a[" Flipkart.com"]+a["Flipkart.com"]
a.drop([" Flipkart.com","Flipkart.com"],axis=1,inplace=True)
a["snapdeal.com"] = a[" Snapdeal.com"]+a["Snapdeal.com"]
a.drop([" Snapdeal.com","Snapdeal.com"],axis=1,inplace=True)

sn.swarmplot(x=a.sum(),y=a.sum(),hue=a.columns)
plt.xlabel('Complete, relevant description information of products')
plt.title("products information", fontsize=15)
plt.show()
```

238 respondents says that amazon gives the complete relevant description information of products, whereas 59 responedents feel that snapdeal and paytm gives the complete relevant description of products which is the least of all.

```python
a = df['Quickness to complete purchase'].str.get_dummies(',')
a["flipkart.com"] = a[" Flipkart.com"]+a["Flipkart.com"]
a.drop([" Flipkart.com","Flipkart.com"],axis=1,inplace=True)
a["paytm.com"] = a[" Paytm.com"]+a["Paytm.com"]
a.drop([" Paytm.com","Paytm.com"],axis=1,inplace=True)
sn.barplot(x=a.sum(),y=a.sum(),hue=a.columns,dodge=False)
plt.xlabel("Quickness to complete purchase")
plt.title("Quickness to complete purchase",fontsize=15)
plt.show()
```

215 respondents believe that amazon has the quickness to complete purchase where snapdeal has the least number of respondents believing that it has quickness to complete purchase

```python
a = df['Security of customer financial information'].str.get_dummies(',')
a["flipkart.com"] = a[" Flipkart.com"]+a["Flipkart.com"]
a.drop([" Flipkart.com","Flipkart.com"],axis=1,inplace=True)
a["paytm.com"] = a[" Paytm.com"]+a["Paytm.com"]
a.drop([" Paytm.com","Paytm.com"],axis=1,inplace=True)
a["myntra.com"] = a[" Myntra.com"]+a["Myntra.com"]
a.drop([" Myntra.com","Myntra.com"],axis=1,inplace=True)
sn.barplot(x=a.sum(),y=a.sum(),hue=a.columns,dodge=False)
plt.xlabel("Security of customer financial information")
plt.title("Security of financial information", fontsize=15)
plt.show()
```

206 respondents believe that amazon.in has the security of customer financial information whereas 88 respondents believe that paytm has the security of financial information which is the least

```python
plt.figure(figsize=(10,5))
a = df['Frequent disruption when moving from one page to another'].str.get_dummies(',')
a["flipkart.com"] = a[" Flipkart.com"]+a["Flipkart.com"]
a.drop([" Flipkart.com","Flipkart.com"],axis=1,inplace=True)
a["snapdeal.com"] = a[" Snapdeal.com"]+a["Snapdeal.com"]
a.drop([" Snapdeal.com","Snapdeal.com"],axis=1,inplace=True)

sn.barplot(x=a.sum(),y=a.sum(),hue=a.columns,dodge=False)
plt.xlabel("Frequent disruption when moving from one page to another")
plt.title("Disruption in website", fontsize=15)
plt.show()
```
